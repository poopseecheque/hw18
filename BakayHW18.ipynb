### 1. Output a list of tickets, where number of flights>3

```python
%%sql $CONNECT_DATA
SELECT ticket_no FROM ticket_flights
GROUP BY ticket_no
HAVING COUNT(flight_id) > 3
LIMIT 1000
```

### 2. Output a total number of tickets which have exactly 5 flights. Just one number, nothing else.

```python
%%sql $CONNECT_DATA
SELECT COUNT(ticket_no) AS total_tickets
FROM (SELECT ticket_no
      FROM ticket_flights
      GROUP BY ticket_no
      HAVING COUNT(flight_id) = 5)
AS tickets_with_5_flights
```

**output the Department_name which has the longest-working employee in the company, together with the First_name + Last_name of this employee.**

```python
%%sql postgresql:///jovyan
SELECT d.DEPARTMENT_NAME, e.FIRST_NAME || ' ' || e.LAST_NAME AS employee_name
FROM Employees e
JOIN Departments d ON e.DEPARTMENT_ID = d.DEPARTMENT_ID
WHERE e.HIRE_DATE = (SELECT MIN(HIRE_DATE) FROM Employees)
```
